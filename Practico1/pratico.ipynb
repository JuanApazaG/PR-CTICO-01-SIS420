```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo CSV
data = pd.read_csv('peso_estatura.csv')

# Visualizar los datos en un grfico de dispersin
plt.scatter(data['Altura'], data['Peso'])
plt.xlabel('Altura (m)')
plt.ylabel('Peso (kg)')
plt.title('Relacin entre Altura y Peso')
plt.show()
# Inicializar valores de w y b
w = 1.0
b = 0.0

# Calcular las predicciones utilizando la frmula y = wx + b
data['Prediccion'] = w * data['Altura'] + b

# Visualizar los datos junto con la recta inicial
plt.scatter(data['Altura'], data['Peso'], label='Datos')
plt.plot(data['Altura'], data['Prediccion'], color='red', label='Recta inicial')
plt.xlabel('Altura (m)')
plt.ylabel('Peso (kg)')
plt.title('Ajuste inicial de la recta')
plt.legend()
plt.show()
```
